#### problema 3
Problema 3: La resta simplePaso 1: Análisis y ResoluciónPlanteamiento: "Un número" ($x$) "al que se le resta siete" ($- 7$) "da como resultado 18" ($= 18$).Ecuación:$$x - 7 = 18$$Despeje: El 7 que está restando pasa al otro lado de la igualdad sumando:$$x = 18 + 7$$Resultado:$$x = 25$$

```python
# Problema 3: Un número menos 7 es igual a 18
# resultado = 18
# resta = 7
resultado = int(input("Ingrese el resultado: "))
resta = int(input("Ingrese la resta: "))

# Despeje: x = 18 + 7
x = resultado + resta

print(f"El número es: {x}")
```

#### Problema 4
Problema 4: Ecuación de primer grado con coeficientePaso 1: Análisis y ResoluciónPlanteamiento: "El triple de un número" ($3x$) "más 4" ($+ 4$) "es igual a 31" ($= 31$).Ecuación:$$3x + 4 = 31$$Despeje:Primero, movemos el 4 que está sumando:$$3x = 31 - 4$$$$3x = 27$$Luego, el 3 que multiplica a la $x$ pasa dividiendo:$$x = \frac{27}{3}$$Resultado:$$x = 9$$

```python
# Problema 4: El triple de un número (3x) + 4 = 31
# total = 31
# termino_independiente = 4
# coeficiente = 3
total = int(input("Ingrese el total: "))
termino_independiente = int(input("Ingrese el termino independiente: "))
coeficiente = int(input("Ingrese el coeficiente: "))

# Paso 1: Restar el 4 -> 3x = 27
# Paso 2: Dividir entre 3 -> x = 9
x = (total - termino_independiente) / coeficiente

print(f"El número es: {int(x)}")
```

#### Problema 5
Problema 5: Perímetro de un rectánguloPaso 1: Análisis y ResoluciónEste requiere entender la geometría del rectángulo. El perímetro es la suma de los cuatro lados ($2 \times \text{largo} + 2 \times \text{ancho}$).Definición de variables:Ancho: $x$Largo: $x + 5$ (mide 5 metros más que el ancho).Planteamiento del Perímetro:$$2(\text{ancho}) + 2(\text{largo}) = 50$$$$2(x) + 2(x + 5) = 50$$Ecuación simplificada:Distribuimos el 2:$$2x + 2x + 10 = 50$$Sumamos las $x$:$$4x + 10 = 50$$Despeje:Restamos 10:$$4x = 50 - 10 \rightarrow 4x = 40$$Dividimos entre 4:$$x = \frac{40}{4} \rightarrow x = 10$$Resultado:Ancho ($x$): 10 metros.Largo ($x+5$): 15 metros.

```python
# Problema 5: Terreno rectangular (Largo = Ancho + 5)
# perimetro_total = 50
# diferencia_largo = 5
perimetro_total = int(input("Ingrese el perímetro total: "))
diferencia_largo = int(input("Ingrese la diferencia de largo: "))
# La ecuación simplificada es 4x + (2 * diferencia) = Perímetro
# 4x + 10 = 50 -> 4x = 40 -> x = 10
ancho = (perimetro_total - (2 * diferencia_largo)) / 4
largo = ancho + diferencia_largo

print(f"El ancho mide: {int(ancho)}m")
print(f"El largo mide: {int(largo)}m")
```
